```python
import math
import seaborn as sns
cols = sub_dataset.columns
n_cols = len(cols)
n_rows = math.ceil(n_cols / 3)  # 3 plots per row (adjust as needed)

fig, axes = plt.subplots(n_rows, 3, figsize=(15, 5 * n_rows))
axes = axes.flatten()

for i, col in enumerate(cols):
    sns.boxplot(y=sub_dataset[col], ax=axes[i])
    axes[i].set_title(col)
```

```python
for col in sub_dataset.columns:
    Q1 = sub_dataset[col].quantile(0.25)
    Q3 = sub_dataset[col].quantile(0.75)
    
    IQR = Q3-Q1
    
    lower_band = Q1 - 1.5*IQR
    upper_band = Q3 + 1.5*IQR
    
    mean_no_outliers = sub_dataset[(sub_dataset[col] <=  upper_band) & (sub_dataset[col] >= lower_band)][col].mean()
    sub_dataset[col] = sub_dataset[col].apply(lambda x: mean_no_outliers if x < lower_band or x > upper_band else x)
```

```python
fig, axes = plt.subplots(n_rows, 3, figsize=(15, 5 * n_rows))
axes = axes.flatten()

for i, col in enumerate(cols):
    sns.boxplot(y=sub_dataset[col], ax=axes[i])
    axes[i].set_title(col)
```

```python
from scipy import stats
for att in sub_dataset:
    mean = sub_dataset[att].mean()
    sub_dataset[att].fillna(mean,inplace=True)
coefs=[]
for att in sub_dataset : 
   
        print('Call Drop Rate  vs',att)
        corr_coef , p_value = stats.pearsonr(sub_dataset['Call Drop Rate'] , sub_dataset[att])
        print(f"Pearson Corr : {corr_coef}")
        print(f"P-value : {p_value}")
        print("###################################")
        coefs.append(corr_coef)

plt.figure()
plt.scatter(coefs,sub_dataset.columns)
plt.show()
```
