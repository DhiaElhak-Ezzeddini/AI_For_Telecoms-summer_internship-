```python
# telecom_kpi_analysis.py
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# =============================
# 1. THROUGHPUT & TRAFFIC KPIs
# =============================
df['DL_Throughput_Mbps'] = (
    df['L.Thrp.bits.DL(bit)'] / (df['L.Thrp.Time.Cell.DL.HighPrecision(ms)'] * 1000)
) * 8 / 1e6

df['UL_Throughput_Mbps'] = (
    df['L.Thrp.bits.UL(bit)'] / (df['L.Thrp.Time.Cell.UL.HighPrecision(ms)'] * 1000)
) * 8 / 1e6

df['DL_Traffic_GB'] = df['L.Thrp.bits.DL(bit)'] / (8 * 1e9)
df['UL_Traffic_GB'] = df['L.Thrp.bits.UL(bit)'] / (8 * 1e9)

# =============================
# 2. RRC SUCCESS RATE
# =============================
rrc_attempts = df[
    ['L.RRC.ConnReq.Att.Emc', 'L.RRC.ConnReq.Att.HighPri', 
     'L.RRC.ConnReq.Att.MoData', 'L.RRC.ConnReq.Att.MoSig', 
     'L.RRC.ConnReq.Att.Mt']
].sum(axis=1)

rrc_success = df[
    ['L.RRC.ConnReq.Succ.Emc', 'L.RRC.ConnReq.Succ.HighPri', 
     'L.RRC.ConnReq.Succ.MoData', 'L.RRC.ConnReq.Succ.MoSig', 
     'L.RRC.ConnReq.Succ.Mt']
].sum(axis=1)

df['RRC_SR_%'] = (rrc_success / rrc_attempts.replace(0, pd.NA)) * 100

# =============================
# 3. E-RAB SETUP SUCCESS RATE
# =============================
df['ERAB_SR_%'] = (df['L.E-RAB.SuccEst'] / df['L.E-RAB.AttEst'].replace(0, pd.NA)) * 100

# =============================
# 4. CALL DROP RATE (CDR)
# =============================
df['CDR_%'] = (
    df['L.E-RAB.AbnormRel'] /
    (df['L.E-RAB.AbnormRel'] + df['L.E-RAB.NormRel']).replace(0, pd.NA)
) * 100

# =============================
# 5. HANDOVER SUCCESS RATE
# =============================
ho_attempts = df[
    ['L.HHO.IntereNB.InterFreq.ExecAttOut', 'L.HHO.IntereNB.IntraFreq.ExecAttOut',
     'L.HHO.IntraeNB.InterFreq.ExecAttOut', 'L.HHO.IntraeNB.IntraFreq.ExecAttOut']
].sum(axis=1)

ho_success = df[
    ['L.HHO.IntereNB.InterFreq.ExecSuccOut', 'L.HHO.IntereNB.IntraFreq.ExecSuccOut',
     'L.HHO.IntraeNB.InterFreq.ExecSuccOut', 'L.HHO.IntraeNB.IntraFreq.ExecSuccOut']
].sum(axis=1)

df['HO_SR_%'] = (ho_success / ho_attempts.replace(0, pd.NA)) * 100

# =============================
# 6. CSFB SUCCESS RATE
# =============================
df['CSFB_SR_%'] = (df['L.CSFB.PrepSucc'] / df['L.CSFB.PrepAtt'].replace(0, pd.NA)) * 100

print(f"CDR_% : \n{df["CDR_%"][:20]}")
```

```python
import pandas as pd
import glob
import os
from collections import defaultdict
from itertools import islice

# --- Setup paths ---
dir_path = "C:\\Users\\DHIA\\Downloads\\OMC\\Sonatel_Huawei_4G"
csv_dir = glob.glob(os.path.join(dir_path, "4G_HUAWEI202506*"))
output_base_dir = os.path.join(dir_path, "per_cell_grouped_output")
os.makedirs(output_base_dir, exist_ok=True)

target_cells = sorted_cell_names[9601:15001]

def chunked(iterable, n):
    it = iter(iterable)
    return iter(lambda: list(islice(it, n)), [])

for group_idx, cell_group in enumerate(chunked(target_cells, 100), start=1):
    group_idx += 96
    print(f" #### Processing group {group_idx} with {len(cell_group)} cells")

    cell_data = defaultdict(list)

    for file_path in csv_dir:
        try:
            df = pd.read_csv(file_path,skiprows=6, sep=',',low_memory=False)

            if 'Time' in df.columns:
                df['Time'] = pd.to_datetime(df['Time'], errors='coerce')

            df = df[df['Cell Name'].isin(cell_group)]

            for cell_name, group in df.groupby("Cell Name"):
                cell_data[cell_name].append(group)

        except Exception as e:
            print(f"Error in file {file_path}: {e}")

    # Save each cell's full data into a file
    ordered_cells_data = []
    for cell_name in cell_group:
        if cell_name in cell_data:
            try:
                df = pd.concat(cell_data[cell_name]).sort_values("Time")
                ordered_cells_data.append(df)
            except Exception as e:
                print(f"Error concatenating data for {cell_name}: {e}")

    if ordered_cells_data:
        group_df = pd.concat(ordered_cells_data)
        output_file = os.path.join(output_base_dir, f"group_{group_idx:03}.csv")
        group_df.to_csv(output_file, index=False)
        print(f"Saved group {group_idx} to: {output_file}")
    else:
        print(f"No data found for group {group_idx}")

print("All groups finished.")
```
